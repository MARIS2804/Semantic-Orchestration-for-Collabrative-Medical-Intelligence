```python
import pandas as pd

# 1. Load the input file
df = pd.read_excel('indicates.xlsx')

# 2. Extract unique values from 'target_node'
# We drop duplicates and also drop any empty (NaN) values to keep the list clean
unique_diseases = df['source_node'].drop_duplicates().dropna()

# 3. Convert to a DataFrame and save to dis.csv
# index=False prevents pandas from adding an extra column for row numbers
unique_diseases.to_csv('sym.csv', index=False)

print(f"Successfully saved {len(unique_diseases)} unique values to sym.csv")
```
